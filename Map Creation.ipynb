```python
#Importing packages
import pgeocode
import pickle
import folium
```

```python
import Postcode Collection.ipynb
```

```python
#Makes a pgeocode object set to Britain's postcodes
nomi = pgeocode.Nominatim('gb')
```

```python
#Calling the codes and names lists saved earlier
codeslist = pickle.load( open("postcodes.p", "rb") )
namelist = pickle.load( open("names.p", "rb") )
```

```python
#Finding the longitudes and latitudes of postcodes in the postcode list
latitudes = []
for postcode in codeslist:
    latitudes.append(nomi.query_postal_code(postcode)['latitude'])

longitudes = []
for postcode in codeslist:
    longitudes.append(nomi.query_postal_code(postcode)['longitude'])
```

```python
#Creating a folium map and placing it on the geographical centre of the UK
centreOfUK = [51.5095, -0.1245]
m = folium.Map(location=[centreOfUK[0],centreOfUK[1]],zoom_tstart=4)
```

```python
#Placing the postcodes as markers on the map and assigning them a label corresponding to that trust's name
for i,j,k in zip(latitudes,longitudes,namelist):
    try:
        folium.Marker(
    location=[i, j],
    popup=str(k)
).add_to(m)
    except:
        pass
        print('error due to:',i,j) #Throws up coordinates that cause errors
```

```python
#Saving the map as an html file
m.save('uk.html')
```
